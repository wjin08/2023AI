```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/admit.txt', names=['ex1','ex2','Admitted'])
print(data)
X=np.c_[data['ex1'], data['ex2']]
y=data['Admitted']
m=len(data)
print(X.shape, y.shape)
```

```python
pos=[]
neg=[]

for(i,val) in enumerate(y):
  if val==1:
    pos.append(i)
  else:
    neg.append(i)

print(pos)
print(neg)
```

```python
import matplotlib.pyplot as plt
plt.plot(X[pos,0].reshape(-1),'b+',label='Admitted')
plt.plot(X[neg,0].reshape(-1),'ro',label='Not Admitted')
plt.xlabel("Exam 1 score")
plt.ylabel("Exam 2 score")
plt.legend(loc='upper right')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='liblinear',C=10)
log_reg.fit(X,y)
```

```python
log_reg.predict([[30,70],[50,90]])
```

```python
x_min, x_max=X[:,0].min(), X[:,0].max()
y_min, y_max=X[:,1].min(), X[:,0].max()
h=.2
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
Z=log_reg.predict(np.c_[xx.ravel(),yy.ravel()])

Z=Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Paired)

plt.plot(X[pos,0].reshape(-1),X[pos,1].reshape(-1),'b+', label='Passed')
plt.plot(X[neg,0].reshape(-1),X[neg,1].reshape(-1),'ro', label='Not Passed')
plt.xlabel("Microchip Test 1")
plt.ylabel("Microchip Test 2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
