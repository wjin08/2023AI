```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Colab Notebooks/sample1.snd'

audio, sr = librosa.load(file_path)
spectrogram = np.abs(librosa.stft(audio))
spectrogram_sizes = np.sum(spectrogram, axis=0)

max_frequency_index = np.argmax(spectrogram_sizes)
max_frequency = librosa.fft_frequencies()[max_frequency_index]

print("스펙트럼 크기:", spectrogram.shape)
print("최대 크기를 가지는 주파수:", max_frequency)

plt.figure(figsize=(12, 8))
librosa.display.specshow(librosa.amplitude_to_db(spectrogram, ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
